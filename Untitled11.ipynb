```python
import csv
import requests
from bs4 import BeautifulSoup

data_list = []

for j in range(1, 10):
    webpage = requests.get('https://www.screener.in/screens/57601/coffee-can-portfolio/?page={}'.format(j)).text
    soup = BeautifulSoup(webpage, 'lxml')

    for i in soup.find_all('td'):
        data_list.append(i.text.strip())

# Reshape the data into rows and columns
num_columns = 14  # Number of columns in your data
num_rows = len(data_list) // num_columns
data_matrix = [data_list[i:i + num_columns] for i in range(0, len(data_list), num_columns)]

# Save data to CSV file
csv_filename = 'stock.csv'
with open(csv_filename, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    for row in data_matrix:
        csv_writer.writerow(row)

print(f'Data saved to {csv_filename}')
```

```python
import csv
import numpy as np
import requests
from bs4 import BeautifulSoup

data_list = []

for j in range(1, 19):
    webpage = requests.get('https://www.screener.in/screens/97687/capacity-expansion/?page={}'.format(j)).text
    soup = BeautifulSoup(webpage, 'lxml')

    for i in soup.find_all('td'):
        data_list.append(i.text.strip())

# Replace empty strings with NaN
data_list = [np.NaN if val == '' else val for val in data_list]

# Reshape the data into rows and columns
num_columns = 14
num_rows = len(data_list) // num_columns
data_matrix = [data_list[i:i + num_columns] for i in range(0, len(data_list), num_columns)]

# Save data to CSV file
csv_filename = 'stock1.csv'
with open(csv_filename, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    for row in data_matrix:
        csv_writer.writerow(row)

print(f'Data saved to {csv_filename}')
```
